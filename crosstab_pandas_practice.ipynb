```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from geopy.geocoders import Nominatim
from time import sleep
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

### 1.How does the average total_bill vary across different days and times of the week?

```python
pd.crosstab(index=df['day'],columns=df['time'],values=df['total_bill'],aggfunc='mean').fillna(0)
```

### 2. What is the relationship between sex and smoker status in terms of the average tip given?

```python
pd.crosstab(index=df['sex'],columns=df['smoker'],values=df['tip'],aggfunc='mean')
```

### 3. How does the average total_bill and tip change with different party sizes (size) on different days?

```python
df.pivot_table(index='day',columns='size',values=['total_bill','tip'],aggfunc='mean').fillna(0)
```
